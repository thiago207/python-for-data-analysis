```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df_iris = pd.DataFrame({
    'comprimento_sepala': [5.1, 7.0, 6.3, 5.8, 6.1],
    'largura_sepala': [3.5, 3.2, 3.3, 2.7, 2.9],
    'comprimento_petala': [1.4, 4.7, 6.0, 5.1, 4.7],
    'largura_petala': [0.2, 1.4, 2.5, 1.9, 1.4],
    'especie': ['setosa', 'versicolor', 'virginica', 'virginica', 'versicolor']
})

display(df_iris)

sns.histplot(x='comprimento_petala', data=df_iris, hue='especie', multiple='stack')
plt.title('Distribuição do Comprimento da Pétala por Espécie')
plt.xlabel('Comprimento da Pétala (cm)')
plt.ylabel('Frequência')
plt.show()
```

```python
df_titanic = pd.DataFrame({
    'Nome': ['Ana', 'Carlos', 'João', 'Maria', 'Pedro'],
    'Idade': [22, 35, 28, 19, 40],
    'Sexo': ['Feminino', 'Masculino', 'Masculino', 'Feminino', 'Masculino'],
    'Classe': [3, 1, 2, 3, 1],
    'Sobreviveu': [0, 1, 1, 1, 0]
})

plt.style.use(style='seaborn-v0_8-whitegrid')
plt.figure(figsize=(8, 4))


df_media_idade_por_sexo = df_titanic.groupby('Sexo')['Idade'].mean().reset_index()
df_media_idade_por_sexo = df_media_idade_por_sexo.set_index('Sexo', drop=True)

bars = plt.barh(df_media_idade_por_sexo.index, df_media_idade_por_sexo['Idade'], height=0.5)

bars[1].set_color("#5d7ed8")
bars[0].set_color("#644aaa")

plt.title('Media de idade por sexo', fontweight='bold', fontsize=25, color = "#8f8f8fe1")    

plt.yticks(fontweight='bold', fontsize=20)

plt.xticks([])
plt.box(False)
plt.tick_params('y',  length=0)

idade = df_media_idade_por_sexo['Idade']
plt.bar_label(bars, labels=[f'{valor:.1f}' for valor in idade], fontweight='bold', fontsize=20, padding=-50)

```
