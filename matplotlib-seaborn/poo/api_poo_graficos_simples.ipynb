# Trabalhando com eixos



## Gráfico de linhas

```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]
```

```python
fig, ax = plt.subplots()
print(fig, ax)

ax.plot(meses, vendas)

ax.set_xlabel('Meses')
ax.set_ylabel('Vendas (milhares de reais)')
ax.set_title('Evolucao vendas')

ax.grid(True)

plt.show()
```

## Gráfico de barras

```python
import matplotlib.pyplot as plt

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]
```

```python
fig, ax = plt.subplots()

ax.bar(produtos, quantidades)


ax.set_xlabel('Produto')
ax.set_ylabel('Quantidades')
ax.set_title('Quantidades de vendas por produtos')


plt.show()
```

## Gráfico de pizza

```python
categorias = ['Eletrônicos', 'Roupas', 'Alimentos', 'Livros']
quantidades = [45, 30, 15, 10]
destaque = [0.1, 0, 0, 0]  # destacar a primeira fatia
```

```python
fig, ax = plt.subplots()

ax.pie(quantidades, labels= categorias, explode=destaque, autopct='%1.1f', shadow=True)

plt.show()
```

## Gráfico de barras empilhadas

```python
anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]
```

```python
fig, ax = plt.subplots()

barra1 = ax.bar(anos, vendas_online, label='Online' ,edgecolor='white')
barra2 = ax.bar(anos, vendas_loja, bottom=vendas_online, label='Loja Fisica', edgecolor='white')

ax.set_xticks(anos)

ax.bar_label(barra1, labels=vendas_online, padding=-15, fontweight='bold', fontsize=12)
ax.bar_label(barra2, labels=vendas_loja, padding=-15, fontweight='bold', fontsize=12)

ax.set_yticks([])


ax.set_ylabel('Vendas (em milhoes de reais)')
ax.set_title('Loja online X Loja Fisica')


plt.legend()
plt.show()
```
