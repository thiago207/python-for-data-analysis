# Trabalhando com eixos



## Mais de um eixo y em uma figura

Crie uma figura de um gráfico de precipitação com um eixo secundário para a temperatura, para mostrar a relação entre a temperatura e a precipitação em uma cidade.



```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
precipitacao_media = [200, 180, 160, 100, 60, 30, 30, 35, 60, 100, 160, 200]
temperatura_media = [30, 29, 27, 25, 23, 22, 22, 23, 24, 25, 26, 28]
```

```python
import matplotlib.pyplot as plt


cor_temperatura = 'tab:red'
cor_percipitacao = 'tab:blue'

fig, ax1 = plt.subplots(figsize=(8,5))



ax1.bar(meses, precipitacao_media, color=cor_percipitacao, label='Percipitacao media (mm)')
ax1.set_ylabel('Percipitacao media (mm)', color=cor_percipitacao)
ax1.tick_params(axis='y', labelcolor=cor_percipitacao)
ax1.set_title('Temperatura e Precipitacao ao longo do ano')



ax2 = ax1.twinx()

ax2.plot(meses, temperatura_media, linewidth=2.5, color=cor_temperatura, label='Temperatura media (C)')
ax2.set_ylabel('Temperatura media (C)', color=cor_temperatura)
ax2.tick_params(axis='y', labelcolor=cor_temperatura)
ax2.set_ylim(20, 30)



fig.legend(bbox_to_anchor=(0.8, 0.85))      
plt.show()
```
