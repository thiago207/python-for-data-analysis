## Gráfico de barras empilhadas para mostrar a distribuição de diferentes tipos de vendas ao longo do tempo

Considere que você tenha uma empresa e queira mostrar a distribuição de diferentes tipos de vendas ao longo do tempo. Crie um gráfico de barras empilhadas:

```python
import matplotlib.pyplot as plt

anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]  #milhoes de reais 
vendas_loja = [10, 8, 6, 4]  #milhoes de reais 

plt.bar(anos, vendas_online, label= 'Online')
plt.bar(anos, vendas_loja, bottom=vendas_online, label = 'Loja Fisica')

plt.xticks(ticks=anos)

plt.title('Vendas Online X Lojas Fisicas')
plt.ylabel('Vendas (em milhoes de reais) ')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]  #milhoes de reais 
vendas_loja = [10, 8, 6, 4]  #milhoes de reais 

cores = plt.get_cmap('Pastel2').colors

plt.bar(anos, vendas_online, label= 'Online', color=cores[2])
plt.bar(anos, vendas_loja, bottom=vendas_online, label = 'Loja Fisica', color=cores[1])

plt.xticks(ticks=anos)

plt.title('Vendas Online X Lojas Fisicas')

plt.ylabel('Vendas (em milhoes de reais) ')

plt.legend()
plt.show()
```
