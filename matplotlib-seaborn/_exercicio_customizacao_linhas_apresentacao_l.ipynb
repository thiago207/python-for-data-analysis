```python
# listas de anos e refeições
anos = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
refeicoes = [40139, 127020, 168193, 153115, 202102, 232897, 277912, 205350, 233389, 232797, 110023, 160350, 222789]
```

```python
import matplotlib.pyplot as plt 
import numpy as np

cor = plt.get_cmap('Set3').colors

plt.figure(figsize=(14, 6))


plt.plot(anos, refeicoes, color=cor[0], linewidth=3,  zorder=1) 
plt.scatter(anos, refeicoes, label='Pontos', color='red', s=100, marker='o', zorder=2)

plt.xticks(anos)

plt.axvspan(2020, 2021, alpha=0.3, color='darkred', zorder=0)
plt.text(2020.03, 60000, 'PANDEMIA', weight='bold')

plt.yticks([])

#pandemia = np.where()
#for i, dados in enumerate(pandemia):
    #print(i, dados)

plt.title('Refeições servidas por ano', fontsize=16, weight='bold', loc='left', pad=20, color='gray')
plt.box(False)
plt.tick_params('x', length=0)
plt.show()
```
