## Gráfico de pizza para mostrar a distribuição de categorias de produtos em uma loja

Suponha que você tenha uma loja e queira mostrar a distribuição de categorias de produtos em sua loja. Você pode usar um gráfico de pizza para mostrar a distribuição de categorias de produtos em sua loja:


O `autopct` é usado para mostrar os valores percentuais. O `%1.1f%%` é usado para mostrar um dígito após o ponto decimal.

```python
import matplotlib.pyplot as plt

categorias = ['Eletrônicos', 'Roupas', 'Alimentos', 'Livros']
quantidades = [45, 30, 15, 10]  # valores em porcentagem
destaques = [0.01, 0, 0, 0.05]

plt.pie(quantidades, labels=categorias, autopct='%.1f%%', explode=destaques)

plt.title('Grafico de pizza', fontweight = 'bold')

plt.show()
```

Mudando o esquema de cores: https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
import matplotlib.pyplot as plt

categorias = ['Eletrônicos', 'Roupas', 'Alimentos', 'Livros']
quantidades = [45, 30, 15, 10]  # valores em porcentagem
destaques = [0.01, 0, 0, 0.05]

cores = plt.get_cmap('tab20').colors

plt.pie(quantidades, labels=categorias, autopct='%.1f%%', explode=destaques, colors=cores)

plt.title('Grafico de pizza', fontweight = 'bold')
plt.show()
```
