## Gráfico de barras para apresentação

Considere que você tem as quantidades de produtos vendidos em uma loja. Faça um gráfico de barras horizontais.

```python
import matplotlib.pyplot as plt

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]
```

```python
barras = plt.barh(produtos, quantidades)
plt.bar_label(barras, labels=quantidades, fontsize=10, fontweight='bold', padding=3)


plt.box(False)
plt.tick_params(axis='y', length=0)
plt.xticks([])

plt.title('Quantidade de produtos vendidos')

plt.show()
```

mais alteraçoes


```python
barras = plt.barh(produtos, quantidades, color='skyblue')
sonamaior = [valor if valor == max(quantidades) else '' for valor in quantidades]


plt.bar_label(barras, labels=sonamaior, fontsize=10, fontweight='bold', padding=3)

cores = plt.get_cmap('Pastel1').colors

plt.box(False)
plt.tick_params(axis='y', length=0)
plt.xticks([])

plt.title('Quantidade de produtos vendidos')

for i, barra in enumerate(barras):
    if i != quantidades.index(max(quantidades)):
        barras[i].set_color(cores[0])
        barras[i].set_alpha(0.9)


plt.show()
```

```python

dados_ordenado = sorted(zip(quantidades, produtos))

produtos_ordenados = [dado[1] for dado in dados_ordenado]
quantidades_ordenadas = [dado[0] for dado in dados_ordenado]

print(quantidades_ordenadas)
print(produtos_ordenados)


barras = plt.barh(produtos_ordenados, quantidades_ordenadas, color='skyblue')
sonamaior = [valor if valor == max(quantidades_ordenadas) else '' for valor in quantidades_ordenadas]


plt.bar_label(barras, labels=sonamaior, fontsize=10, fontweight='bold', padding=3)

cores = plt.get_cmap('Pastel1').colors

plt.box(False)
plt.tick_params(axis='y', length=0)
plt.xticks([])

plt.title('Quantidade de produtos vendidos')

for i, barra in enumerate(barras):
    if i != quantidades_ordenadas.index(max(quantidades_ordenadas)):
        barras[i].set_color(cores[1])
        barras[i].set_alpha(0.9)


plt.show()
```
