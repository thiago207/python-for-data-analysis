## Gráfico de barras empilhadas para apresentação 

```python
import matplotlib.pyplot as plt
from cycler import cycler

anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

plt.figure(figsize=(8,4))
# rc = runtime configuration (configuração quando executa)
cores = plt.get_cmap('Pastel2').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)
 
plt.bar(anos, vendas_online, label='Online')
plt.bar(anos, vendas_loja, bottom=vendas_online, label='Loja')

# forçar os anos a serem exibidos como inteiros
plt.xticks(ticks=anos)

plt.xlabel('Ano')
plt.ylabel('Vendas (em milhões)')
plt.title('Vendas Online vs Loja')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from cycler import cycler

anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

plt.figure(figsize=(8,6))
# rc = runtime configuration (configuração quando executa)
cores = plt.get_cmap('Pastel2').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)
 
barras_vendas_on = plt.bar(anos, vendas_online, label='Online')
barras_loja_fisicas = plt.bar(anos, vendas_loja, bottom=vendas_online, label='Loja')

# forçar os anos a serem exibidos como inteiros
plt.xticks(ticks=anos)

plt.bar_label(barras_loja_fisicas, labels=vendas_loja, fontweight='bold',label_type='center')
plt.bar_label(barras_vendas_on, fontweight='bold', labels=vendas_online, label_type='center')

plt.box(False)
plt.tick_params(axis='x', length=0)


plt.ylabel('Vendas (em milhões)')
plt.title('Vendas Online vs Loja')
plt.legend()
plt.show()
```
