## Gráfico de linha para apresentação


```python
import matplotlib.pyplot as plt



meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

plt.plot(meses, vendas)

plt.annotate('Menor valor', color='red', xy=(2, 78), xytext=(3, 80), arrowprops=dict(facecolor='red'))

plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

marcadores = np.where(vendas == np.min(vendas), 'x', '')

plt.plot(meses, vendas)

for posicao, marcador in enumerate(marcadores):
    if marcador == 'x':
        cor='red'
    else:
        cor='blue'
    plt.scatter(meses[posicao], vendas[posicao], marker=marcador, color=cor, s=100)

plt.annotate('Menor valor', color='red', xy=(2, 78), xytext=(2.35, 78))

plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

marcadores = np.where(vendas == np.min(vendas), 'x', '')


#grafico
plt.plot(meses, vendas)

#marcadores
for posicao, marcador in enumerate(marcadores):
    if marcador == 'x':
        cor='red'
    else:
        cor='blue'
    plt.scatter(meses[posicao], vendas[posicao], marker=marcador, color=cor, s=100)

#anotaçao
plt.annotate('Menor valor', color='red', xy=(2, 78), xytext=(2.35, 78))

#destaque de regiao
plt.axvspan(6, 11, alpha=0.4, color='green')
plt.text(6.25, 80,'Recupecao crescimento', weight='bold')



plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

marcadores = np.where(vendas == np.min(vendas), 'x', '')


#grafico
plt.plot(meses, vendas)

#marcadores
for posicao, marcador in enumerate(marcadores):
    if marcador == 'x':
        cor='red'
    else:
        cor='blue'
    plt.scatter(meses[posicao], vendas[posicao], marker=marcador, color=cor, s=100)
    plt.annotate(vendas[posicao],color=cor, xy=(meses[posicao], vendas[posicao]), xytext=(-5, 5), textcoords='offset points')

#anotaçao
#plt.annotate('Menor valor', color='red', xy=(2, 78), xytext=(2.35, 78))

#destaque de regiao
plt.axvspan(6, 11, alpha=0.4, color='green')
plt.text(6.25, 80,'Recupecao crescimento', weight='bold')

plt.box(False)
plt.tick_params(axis='x', length=0) #tirar os traços
plt.yticks([])


plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano')
plt.show()
```
